```python
import pandas as pd

# Define the file path (replace with your actual path)
file_path = "C:\\Users\\syash\\Downloads\\DS_ML Coding Challenge Dataset (1).xlsx - Training Dataset.csv"

# Read the data from the CSV file
train_data = pd.read_csv(file_path)

# Remove rows with missing values in the 'Cost' column
train_data.dropna(subset=['Sourcing Cost'], inplace=True)

# Target variable (assuming 'Cost' is the target)
y = train_data['Sourcing Cost']

# Remove target variable from training data
train_data.drop('Sourcing Cost', axis=1, inplace=True)

# Handle columns with missing values (replace with your desired strategy)
# This example removes columns with any missing values
# You can choose to impute missing values or use other techniques

# train_data.dropna(axis=1, inplace=True)  # Remove entire columns with missing values (alternative)

# Select numeric columns only
numeric_cols = [cname for cname in train_data.columns if train_data[cname].dtype in ['int64', 'float64']]
X = train_data[numeric_cols].copy()

print("Shape of input data: {} and shape of target variable: {}".format(X.shape, y.shape))

X.head()  # Show first 5 training examples
```

```python
# Lets split the data into 5 folds.  
# We will use this 'kf'(KFold splitting stratergy) object as input to cross_val_score() method
kf =KFold(n_splits=5, shuffle=True, random_state=42)

cnt = 1
# split()  method generate indices to split data into training and test set.
for train_index, test_index in kf.split(X, y):
    print(f'Fold:{cnt}, Train set: {len(train_index)}, Test set:{len(test_index)}')
    cnt += 1
```

```python
"""
Why we are using '-' sign to calculate RMSE?
ANS: Classification accuracy is reward function, means something you want to maximize. Mean Square Error is loss function, 
means something you want to minimize. Now if we use 'cross_val_score' function then best score(high value) will give worst 
model in case of loss function! There are other sklearn functions which also depends on 'cross_val_score' to select best model by
looking for highest scores, so a design decision was made for 'cross_val_score' to negate the output of all loss function. 
So that when other sklearn function calls 'cross_val_score' those function can always assume that highest score indicate better model.
In short ignore the negative sign and rate the error based on its absolute value.
"""
def rmse(score):
    rmse = np.sqrt(-score)
    print(f'rmse= {"{:.2f}".format(rmse)}')
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold, train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression

# Read data from CSV file (handle potential errors)
try:
  data_path = "C:/Users/syash/Downloads/DS_ML Coding Challenge Dataset (1).xlsx - Training Dataset.csv"
  df = pd.read_csv(data_path)
except FileNotFoundError:
  print("Error: File not found. Please check the path and try again.")
  exit()

# Encode categorical features
le = LabelEncoder()
for col in df.columns:
  if df[col].dtype == object:  # Check for string type
    df[col + " Encoded"] = le.fit_transform(df[col])
    df.drop(col, axis=1, inplace=True)  # Drop original categorical column

# Separate features and target variable
X = df.drop("Sourcing Cost", axis=1)  # Features
y = df["Sourcing Cost"]  # Target variable

# Define K-Fold Cross-Validation
kfold = KFold(n_splits=10)  # Adjust k as needed

# Create lists to store evaluation metrics (optional)
rmse_scores = []  # Root Mean Squared Error

# K-Fold loop
for train_index, test_index in kfold.split(X):
  X_train, X_test = X.iloc[train_index], X.iloc[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  # Define and train the model
  model_regression = LinearRegression()
  model_regression.fit(X_train, y_train)

  # Make predictions on the test set
  y_pred = model_regression.predict(X_test)

  # Calculate Root Mean Squared Error (optional)
  from sklearn.metrics import mean_squared_error
  rmse = mean_squared_error(y_test, y_pred, squared=False)  # Square root of MSE
  rmse_scores.append(rmse)

# Print K-Fold results (optional)
if len(rmse_scores) > 0:
  print("K-Fold Cross-Validation Results (RMSE):")
  print(f"Mean RMSE: {sum(rmse_scores) / len(rmse_scores):.2f}")
  print(f"Standard Deviation: {np.std(rmse_scores):.2f}")  # Import NumPy for std

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
 
def removal_box_plot(df, column, threshold):
    sns.boxplot(df[column])
    plt.title(f'Original Box Plot of {column}')
    plt.show()
 
    removed_outliers = df[df[column] <= threshold]
 
    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers
 
 
threshold_value = 100
 
no_outliers = removal_box_plot(df_Product, 'Sourcing Cost', threshold_value)
```
